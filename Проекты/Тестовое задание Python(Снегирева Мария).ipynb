# Задание 3.2
Рассчитайте результаты АВ-теста и с помощью статистического критерия сделайте вывод

```python
def test_calc(r1, r2, alpha = 0.05):
    
    s,p = ttest_ind(r1, r2)
    
    if p < alpha:
        print("Гипотеза Н0 не подтверждается: средние не равны")
        print("Разница между средними = ", r1.mean() - r2.mean())
    else:
        print("Гипотеза Н0 подтверждается: средние равны")
        print("Разница между средними = 0")
        
    return s, p
```

```python
test_calc(df_dataset_city[df_dataset_city['id_group'] == 1]['nflag_order'], df_dataset_city[df_dataset_city['id_group'] == 0]['nflag_order'])
```
